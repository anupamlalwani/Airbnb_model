## Multivariate Linear Model

```python
#handling null values
X.bathrooms = X.bathrooms.fillna(1)
X.bedrooms=X.bedrooms.fillna(1)
```

```python
X.isnull().sum()
```

```python
#handling special characters in price column and de
y = oct_df['price'].replace({'\$':'',',':''}, regex=True)
y = y.astype(np.float)
y = y.values.reshape(-1,1)
X.shape, y.shape
# len(y)
```

```python
#splitting dataset into training and test data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=150)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
#creating model and trying fit

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
model
```

```python
#finding model coeffs and intercept
model.coef_[0], model.intercept_
```

```python
# #predicting price using model

pred = model.predict(X_test)
pred
```

```python
#plotting predictions and test values
import matplotlib.pyplot as plt
plt.xlim(0,2000)
plt.ylim(0,2000)
plt.plot(y_test,y_test, c= 'blue')
plt.scatter(y_test, pred, c='green')
plt.show()
```

```python
#model error and score
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test,pred)
r2 = r2_score(y_test,pred)
mse, r2
```

## Model 3 - Gradient Boosting using Amenities

```python
#import model dependencies
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn.metrics import mean_absolute_error,r2_score
from sklearn.externals import joblib
```

```python
# Fit regression model
model3 = ensemble.GradientBoostingRegressor(
    n_estimators=3000,
    learning_rate=0.01,
    max_depth=3,
    min_samples_leaf=20,
    max_features=2,
    loss='ls',
    random_state=200  
)
```

```python
model3.fit(X_train,y_train)
```

```python
# Find the error rate on the training set

mse = mean_absolute_error(y_train, model3.predict(X_train))
print("Training Set Mean Absolute Error: %.4f" % mse)

# Find the error rate on the test set
mse = mean_absolute_error(y_test, model3.predict(X_test))
print("Test Set Mean Absolute Error: %.4f" % mse)

#r2 score
pred2 = model3.predict(X_test)
r2 = r2_score(y_test, pred2)
print("r2 score: %.4f" % r2)
```
